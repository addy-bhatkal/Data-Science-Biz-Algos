```python
df = pd.read_csv('Groceries_dataset.csv')
df.head()
```

```python
df['itemDescription'] = df['itemDescription'].str.strip()
df
```

```python
df['Member_number'].apply(type).unique()
```

```python
basket = df.groupby(['Member_number','itemDescription'])['Date'].sum().unstack().reset_index().fillna(0).set_index('Member_number')
basket
```

```python
def encode(x):
    if x == 0:
        return 0
    if x != 0:
        return 1
```

```python
basket = basket.applymap(encode)
basket
```

```python
freq = apriori(basket, min_support=0.05, use_colnames= True)
freq
```

```python
rules = association_rules(freq, min_threshold=1, metric = 'lift')
rules
```
