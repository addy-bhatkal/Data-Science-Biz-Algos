```python
movies1 = pd.read_csv('C:\\Users\\ABC\\6. Datasets\\Movie reccomender Movie_Titles_Id1.csv', names = ['movie id', 'movie title'])
movies2 = pd.read_csv('C:\\Users\\ABC\\6. Datasets\\Recommend.csv', names = ['user id', 'movie id', 'rating', 'timestamp'])

movies2.head()
```

```python
df = pd.merge(movies1, movies2, on = 'movie id')
df.head()
```

```python
df.groupby('movie title')['rating'].mean().sort_values(ascending=False)
```

```python
df.groupby('movie title')['rating'].count().sort_values(ascending=False)
```

```python
ratings = pd.DataFrame(df.groupby('movie title')['rating'].mean().sort_values(ascending=False))
ratings['nofratings'] = pd.DataFrame(df.groupby('movie title')['rating'].count().sort_values(ascending=False))
ratings
```

```python
ratings['nofratings'] = pd.DataFrame(df.groupby('movie title')['rating'].count())
ratings['nofratings'].sort_values(ascending=False)
```

# Part II

```python
moviemat = df.pivot_table(index = 'user id', columns = 'movie title', values = 'rating')
moviemat.head()
```

```python
starwarsratings = moviemat['Star Wars (1977)']
```

```python
SIMsw = moviemat.corrwith(starwarsratings)
SIMsw.sort_values()

SIMsw2 = pd.DataFrame(SIMsw, columns = ['Correlation'])
SIMsw2['Correlation'].sort_values(ascending = False)

```

```python
SIMsw2 = SIMsw2.join(ratings['nofratings'])

```

```python
SIMsw2.sort_values('nofratings', ascending=False)
```
