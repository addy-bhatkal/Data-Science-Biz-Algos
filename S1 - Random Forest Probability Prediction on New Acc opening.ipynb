```python
pb = pd.read_csv('C:\\Users\\ABC\\6. Datasets\\bank_cleaned.csv')
pb.head()
```

```python
pb['default'] = pb['default'].map({'no': 0, 'yes': 1, 'unknown': 0})
pb['response'] = pb['response'].map({'no':0, 'yes': 1})
pb.head()
```

```python
x = pb[['age', 'default', 'balance', 'duration']]
y = pb['response']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train, y_train)
```

```python
y_pred = rf.predict(x_test)
y_pred
```

```python
pb['datatype'] = 'train'
pb.loc[pb.index.isin(x_test.index), 'datatype'] = 'test'
pb.head()
```

```python
final = pd.DataFrame({'y_test': y_test, 'y_pred': y_pred})
final
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_test, y_pred)
acc
```
